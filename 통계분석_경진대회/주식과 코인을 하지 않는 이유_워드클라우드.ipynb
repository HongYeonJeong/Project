# **주식, 코인 안하는 이유 워드 클라우드**

### 파일 읽기

```python
f = open('주코_안하는_이유.txt', 'r', encoding='utf-8')
j_result = f.read()
print(j_result)
```

### 토크나이즈

```python
from konlpy.tag import Okt

t = Okt()
tokens = t.morphs(j_result)

print('# 토큰 수 : ', len(tokens))
print('# 토큰 예시 : ')
print(tokens[:300])
```

```python
print(t.pos(j_result))
```

### 불용어 제거

```python
my_tag = ['Noun', 'Adjective', 'Verb']
my_words = [word for word, tag in t.pos(j_result) if tag in my_tag]

print(my_words[:500])
```

```python
stop_words = ['그', '잘', '대한', '있구요', '처럼', '보이', '않는', '팔아서', '찾으면', '합니다', '입니다', 
              '반면', '솔직히', '가시', '놓고'] 
my_words = [word for word in my_words if word not in stop_words]

print(my_words)
```

### 빈도수 사전

```python
word_dict = {}

for word in my_words:
    word_dict[word] = word_dict.get(word, 0) + 1
    
sorted_dict = sorted(word_dict, key=word_dict.get, reverse=True)

for key in sorted_dict:
    print(f'{repr(key)} : {word_dict[key]}')
    print()
```

### 워드 클라우드

```python
import numpy as np
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud

from PIL import Image

font_path = 'c:/Windows/Fonts/malgun.ttf'
coin_mask = np.array(Image.open("coin.png"))

wc = WordCloud(
    font_path = font_path,
    max_font_size = 100,
    width = 800,
    height = 400,
    background_color = 'white',
    mask = coin_mask)

wc.generate_from_frequencies(word_dict)     
wc.to_file("j_result.png")

plt.figure()
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
