```python
f = open('투자_안하는_이유.txt', 'r', encoding='utf-8')
t_result = f.read()
print(t_result)
```

```python
from konlpy.tag import Okt

t = Okt()
tokens = t.morphs(t_result)

print('# 토큰 수 : ', len(tokens))
print('# 토큰 예시 : ')
print(tokens[:300])
```

```python
print(t.pos(t_result))
```

```python
my_tag = ['Noun', 'Adjective', 'Verb']
my_words = [word for word, tag in t.pos(t_result) if tag in my_tag]

print(my_words[:500])
```

```python
stop_words = ['하기엔', '입니다', '대한', '가시'] 
my_words = [word for word in my_words if word not in stop_words]
print(my_words)
```

```python
word_dict = {}

for word in my_words:
    word_dict[word] = word_dict.get(word, 0) + 1
    
sorted_dict = sorted(word_dict, key=word_dict.get, reverse=True)

for key in sorted_dict[:100]:
    print(f'{repr(key)} : {word_dict[key]}')
    print()
```

```python
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud

font_path = 'c:/Windows/Fonts/malgun.ttf'

wc = WordCloud(
    font_path = font_path,
    max_font_size = 100,
    width = 800,
    height = 400,
    background_color = 'white',
    max_words = 80)

wc.generate_from_frequencies(word_dict)

plt.figure()
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
